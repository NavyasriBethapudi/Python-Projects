```python
# Remove uneccessary columns from each table.
#hotel_info
hotel.drop("listing_url", axis=1, inplace=True)
hotel.drop("neighborhood_overview", axis=1, inplace=True)
hotel.drop("picture_url", axis=1, inplace=True)
hotel.head()
```

```python
# host_info
host.drop("host_url", axis=1, inplace=True)
host.drop("host_about", axis=1, inplace=True)
host.drop("host_thumbnail_url", axis=1, inplace=True)
host.drop("host_picture_url", axis=1, inplace=True)
host.drop("host_neighbourhood", axis=1, inplace=True)
host.head()
```

```python
#location_info
location.drop("neighbourhood_cleansed", axis=1, inplace=True)
location.head()
```

```python
#reviews_info
review.drop("calculated_host_listings_count", axis=1, inplace=True)
review.drop("calculated_host_listings_count_entire_homes", axis=1, inplace=True)
review.drop("calculated_host_listings_count_private_rooms", axis=1, inplace=True)
review.drop("calculated_host_listings_count_shared_rooms", axis=1, inplace=True)
review.drop("review_scores_accuracy", axis=1, inplace=True)
review.head()
```

```python
# Find the Missing Data from each csv file inserted.
#host info
missing_host = host.isnull()
missing_host.head()

for column in missing_host.columns.values.tolist():
    print(column)
    print(missing_host[column].value_counts())
    print(" ")
```

# Host_Info

```python
#Replace Missing Values with Values 
#host_is_superhost
max_superhost = host["host_is_superhost"].value_counts().idxmax()
host["host_is_superhost"].replace(np.NaN, max_superhost, inplace=True)
host.head()

#host_listings_count
avg_host_list_count = host["host_listings_count"].mean(axis=0)
host["host_listings_count"].replace(np.NaN, avg_host_list_count, inplace=True)
host.head()

#host_total_listings_count
avg_host_totlist_count = host["host_total_listings_count"].mean(axis=0)
host["host_total_listings_count"].replace(np.NaN, avg_host_totlist_count, inplace=True)
host.head()

#host_has_profile_pic
max_hostprofile = host["host_has_profile_pic"].value_counts().idxmax()
host["host_has_profile_pic"].replace(np.NaN, max_hostprofile, inplace=True)
host.head()

#host_has_profile_pic
max_hostidentity = host["host_identity_verified"].value_counts().idxmax()
host["host_identity_verified"].replace(np.NaN, max_hostidentity, inplace=True)
host.head()
```

```python
#host_response_rate
host["host_response_rate"] = host["host_response_rate"].str.rstrip("%").astype("float")/100.00
avg_host_resp_rate = host["host_response_rate"].mean(axis=0)
host["host_response_rate"].replace(np.NaN, avg_host_resp_rate, inplace=True)
host.head()

#host_acceptance_rate
host["host_acceptance_rate"] = host["host_acceptance_rate"].str.rstrip("%").astype("float")/100.00
avg_host_accp_rate = host["host_acceptance_rate"].mean(axis=0)
host["host_acceptance_rate"].replace(np.NaN, avg_host_accp_rate, inplace=True)
host.head()

#host_response_time
max_host_resp_time = host["host_response_time"].value_counts().idxmax()
host["host_response_time"].replace(np.NaN, max_host_resp_time, inplace=True)
host.head()

#host_location
max_host_loc = host["host_location"].value_counts().idxmax()
host["host_location"].replace(np.NaN, max_host_loc, inplace=True)
host.head()
```

```python
# Drop Missing Values
#host_name -> 5 missing values & #host_since -> 5 missing values
host.dropna(subset=["host_name"], axis=0, inplace=True)
host.reset_index(drop=True, inplace=True)
```
